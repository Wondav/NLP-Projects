```python
import sagemaker
import boto3
from sagemaker.huggingface import HuggingFace
from sagemaker.experiments.run import Run
```

```python
sess = sagemaker.Session()
role = sagemaker.get_execution_role()
```

```python
hyperparameters = {
    'model_name_or_path':'t5-small',
    'epochs':1,
    # add your remaining hyperparameters
}

env = {
    'SAGEMAKER_REQUIREMENTS': 'requirements.txt',
}
exp_name = "text-summarization-script-mode-experiment"
```

```python
with Run(experiment_name=exp_name, sagemaker_session=sess) as run:
    run.log_parameter("epochs", hyperparameters['epochs'])
    run.log_parameter("model_name", hyperparameters['model_name_or_path'])
    
    huggingface_estimator = HuggingFace(entry_point='train.py',
                            source_dir='./scripts',
                            instance_type='ml.p3.2xlarge',
                            instance_count=1,
                            role=role,
                            transformers_version='4.17',
                            tensorflow_version='2.6',
                            py_version='py38',
                            env=env,
                            hyperparameters = hyperparameters)
    huggingface_estimator.fit()
```
